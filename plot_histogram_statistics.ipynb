```python
import pandas as pd
import matplotlib.pyplot as plt

#Generates the histograms for the heat pump type distribution and the different combinations of heat pump brand, refrigerant and size

# Load dataset the path where the modified all_summary.csv file is located 
all_summary = ... 

# Filter dataset
all_summary = all_summary[all_summary['Included_SPF_analysis'] == True]

# Plot heat pump type distribution
plt.figure()
all_summary['HP_Installed'].hist()
plt.ylabel('Number of heat pumps')
plt.xlabel('Type of heat pump')
plt.title('Heat pumps type distribution')
plt.grid(True, which='both')

# List of conditions for histograms
conditions = [
    ("HP_Installed == 'ASHP'", 'HP_Brand'),
    ("HP_Installed == 'HT_ASHP'", 'HP_Brand'),
    ("HP_Installed == 'GSHP'", 'HP_Brand'),
    ("HP_Installed == 'Hybrid'", 'HP_Brand'),
    ("HP_Installed == 'ASHP'", 'HP_Refrigerant'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Daikin'", 'HP_Refrigerant'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Mitsubishi'", 'HP_Refrigerant'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Vaillant'", 'HP_Refrigerant'),
    ("HP_Installed == 'HT_ASHP'", 'HP_Refrigerant'),
    ("HP_Installed == 'GSHP'", 'HP_Refrigerant'),
    ("HP_Installed == 'Hybrid'", 'HP_Refrigerant'),
    ("HP_Installed == 'Hybrid' & HP_Brand == 'Daikin'", 'HP_Size_kW'),
    ("HP_Installed == 'GSHP' & HP_Brand == 'Kensa'", 'HP_Size_kW'),
    ("HP_Installed == 'GSHP' & HP_Brand == 'Vaillant'", 'HP_Size_kW'),
    ("HP_Installed == 'HT_ASHP' & HP_Brand == 'Daikin'", 'HP_Size_kW'),
    ("HP_Installed == 'HT_ASHP' & HP_Brand == 'Vaillant'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Vaillant'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Daikin'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Daikin' & HP_Refrigerant == 'R32'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Daikin' & HP_Refrigerant == 'R410A'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Mitsubishi'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Mitsubishi' & HP_Refrigerant == 'R410A'", 'HP_Size_kW'),
    ("HP_Installed == 'ASHP' & HP_Brand == 'Mitsubishi' & HP_Refrigerant == 'R32'", 'HP_Size_kW'),
]

# Generate histograms
for condition, column in conditions:
    plt.figure()
    all_summary.query(condition)[column].hist()
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column} for {condition}')
    plt.grid(True, which='both')

plt.show()
```
